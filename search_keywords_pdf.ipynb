```python
import os
import pdfplumber
import pandas as pd
```

```python
def buscar_termos_em_pdfs(diretorio, termos):
    resultados = []
    
    for arquivo in os.listdir(diretorio):
        if arquivo.lower().endswith(".pdf"):
            caminho_arquivo = os.path.join(diretorio, arquivo)
            print(arquivo)
            try:
                with pdfplumber.open(caminho_arquivo) as pdf:
                    for i, pagina in enumerate(pdf.pages):
                        texto = pagina.extract_text()
                        if texto:
                            for termo in termos:
                                if termo.lower() in texto.lower():
                                    resultados.append({
                                        "Arquivo": arquivo,
                                        "Página": i + 1,
                                        "Termo Encontrado": termo
                                    })
            except Exception as e:
                print(f"Erro ao processar {arquivo}: {e}")
    
    return resultados
```

```python
diretorio_pdfs = "G:/Drives compartilhados/GESTAO/_Operacional/Planilhas Gestão/Scripts/temp/temp_regulamentos/" # Substitua pelo diretório correto
termos_busca = ["Saque Aniversário", "FGTS", "INSS"]  

resultados = buscar_termos_em_pdfs(diretorio_pdfs, termos_busca)
```

```python
# Exportar resultados para um arquivo CSV
if resultados:
    df = pd.DataFrame(resultados)
    df.to_csv("resultados_busca.csv", sep= ';', index=False, encoding="latin1")
    print("Resultados exportados para 'resultados_busca.csv'")
else:
    print("Nenhum termo encontrado.")
```
