```python
base_fundos = pd.read_excel("C:/Users/lucas.andrade_cultin/Downloads/Cópia de emissoes_julho24_emdiante.xlsx", sheet_name= "Quantum Axis")
```

```python
cols = ["CNPJ", "Nome"]
cnpj_fundos = base_fundos.loc[base_fundos["Responsavel"] == "TEDIM/FELIPE", cols]
```

```python
wait = WebDriverWait(navegador, 60)

def wait_xpath(xpath):
    wait.until(EC.element_to_be_clickable((By.XPATH, xpath))).click()
```

```python
for index, row in cnpj_fundos.iterrows():

    doc  = row["CNPJ"]
    name = row["Nome"]

    try:
        cnpj = navegador.find_element(By.NAME, 'filtroRegistroFundoBean.numeroRegistro')
        cnpj.click()
        cnpj.send_keys(Keys.CONTROL + "a")
        cnpj.send_keys(Keys.BACKSPACE)
        cnpj.send_keys(doc)

        time.sleep(8)
        wait_xpath('//*[@id="panelFiltros"]/div/div/a[1]')
        # navegador.find_element(By.XPATH, '//*[@id="panelFiltros"]/div/div/a[1]').click()

        time.sleep(8)
        wait_xpath('//*[@id="containerView"]/div[2]/div[3]/table/tbody/tr/td[8]/a[1]')
        # navegador.find_element(By.XPATH, '//*[@id="containerView"]/div[2]/div[3]/table/tbody/tr/td[8]/a[1]').click()

        time.sleep(12)
        wait_xpath('//*[@id="tabFundo"]/ul/li[3]')
        # navegador.find_element(By.XPATH, '//*[@id="tabFundo"]/ul/li[3]').click()

        time.sleep(15)  
        wait_xpath('//*[@id="tabPanelVisualizarRegulamento"]/div/div[3]/table/tbody/tr/td[1]/input')                
        # navegador.find_element(By.XPATH, '//*[@id="tabPanelVisualizarRegulamento"]/div/div[3]/table/tbody/tr/td[1]/input').click()

        time.sleep(2)
        wait_xpath('//*[@id="tabPanelVisualizarRegulamento"]/div/div[3]/table/tbody/tr/td[4]/div/div/a')
        # navegador.find_element(By.XPATH, '//*[@id="tabPanelVisualizarRegulamento"]/div/div[3]/table/tbody/tr/td[4]/div/div/a').click()

        time.sleep(7)
        wait_xpath('//*[@id="containerView"]/div[3]/div[1]/a')
        # navegador.find_element(By.XPATH, '//*[@id="containerView"]/div[3]/div[1]/a').click()

        time.sleep(5)  # Aguarda o download terminar
        downloaded_file = max([os.path.join(pasta_downloads, f) for f in os.listdir(pasta_downloads)], key=os.path.getctime)
        print(downloaded_file)
        
        new_file_path = os.path.join(pasta_downloads, f"{name}.pdf")
        print(new_file_path)
        os.rename(downloaded_file, new_file_path)
        
    except:
        navegador.find_element(By.XPATH, '//*[@id="containerView"]/div[3]/div[1]/a').click()
        continue

# for arquivo, _ in arquivos_modificados:
#     caminho_origem = os.path.join(pasta_downloads, arquivo)
#     caminho_destino = os.path.join(pasta_fim, arquivo)
#     shutil.move(caminho_origem, caminho_destino)
#     print(f"Arquivo '{arquivo}' movido para '{pasta_fim}'")
```
