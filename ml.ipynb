```python
import json
import os
from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import to_categorical
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Caminhos das pastas
train_image_dir = 'path/to/train_images'
test_image_dir = 'path/to/test_images'
json_file_path = 'path/to/annotations.json'

# Carregar as anotações do JSON
with open(json_file_path, 'r') as file:
    annotations = json.load(file)

# Função para carregar imagens e seus rótulos
def load_data(image_dir, annotations=None):
    images = []
    labels = []

    for image_name in os.listdir(image_dir):
        if image_name.endswith('.jpg') or image_name.endswith('.png'):
            img_path = os.path.join(image_dir, image_name)
            img = image.load_img(img_path, target_size=(128, 128))  # Redimensionando para 128x128
            img_array = image.img_to_array(img) / 255.0  # Normalizando a imagem

            images.append(img_array)

            if annotations:  # Para a base de treino, associamos o rótulo
                texture = annotations.get(image_name)
                labels.append(texture)

    return np.array(images), np.array(labels)

# Carregar dados de treinamento e teste
train_images, train_labels = load_data(train_image_dir, annotations)
test_images, _ = load_data(test_image_dir)

# Convertendo rótulos para categorias numéricas (se necessário)
unique_textures = list(set(train_labels))  # Lista de todas as texturas únicas
texture_to_idx = {texture: idx for idx, texture in enumerate(unique_textures)}
train_labels_numeric = np.array([texture_to_idx[label] for label in train_labels])
train_labels_categorical = to_categorical(train_labels_numeric)

# Divisão treino/teste
X_train, X_val, y_train, y_val = train_test_split(train_images, train_labels_categorical, test_size=0.2, random_state=42)
```
